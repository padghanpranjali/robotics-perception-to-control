```python
import cv2
import matplotlib.pyplot as plt

rgb = cv2.imread("000001-color.png")
rgb = cv2.cvtColor(rgb, cv2.COLOR_BGR2RGB)

depth = cv2.imread("000001-depth.png", cv2.IMREAD_UNCHANGED)

plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
plt.title("RGB Image")
plt.imshow(rgb)
plt.axis("off")

plt.subplot(1,2,2)
plt.title("Depth Image")
plt.imshow(depth, cmap="gray")
plt.axis("off")
plt.show()
```

```python
print("Depth min:", depth.min())
print("Depth max:", depth.max())
print("Depth dtype:", depth.dtype)
```

Depth images encode distance from the camera optical center and are noisy near object boundaries.
